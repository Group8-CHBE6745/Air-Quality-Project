```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# Data Input
Read and format the data from csv file


```python
# Read the data from csv
df = pd.read_csv('data/AirQualityUCI.csv', sep = ';', decimal = ',')

# Show 
df.head()
```

```python
# The last two unnamed columns need to be dropped
df.drop(columns = ['Unnamed: 15', 'Unnamed: 16'], axis = 1, inplace = True)
# Create header list
headers = [header for header in df]
print(headers)
# Show
df.head()
```

```python
# Make sure all the datatype should be float64 except Date & Time
df.dtypes
```

# Data Cleaning

```python
# Show the end of the df to see if there is any empty value
df.tail()
```

```python
# Drop end rows with NaN values
df.dropna(how = 'all', inplace = True)
# Show
df.tail()
```

```python
# Check if each attribute has the same count of data
print('Count of each values:\n', df.count())
```

From the data decription, the missing values in the dataset are already replaced by -200. However, it is better to make sure there are no Nan and negative values in the dataset.

```python
# Now, we are looking for the nagetive values and Nan
def invalid_check(headers: list)->None:
    nan_list = [0] * 15
    print(f"Count of Negative Value or NaN out of {df.shape[0]} rows")
    for i in range(len(headers)):
        for data in df[headers[i]]:
            # looking for the nagetive value and Nan except data and time
            if data == float and (data == None or data < 0):
                nan_list[i] += 1
        print(f"{headers[i]:<13}    {nan_list[i]:>4}")
invalid_check(headers)
```

Now, we will search for -200 values, as these are considered missing values according to the data description.

```python
# Now, we are looking for the missing data in the dataframe
def missing_search(headers: list)->None:
    nan_list = [0] * 15
    print(f"Count of missing value out of {df.shape[0]} rows")
    for i in range(len(headers)): #Loop over each column
        for data in df[headers[i]]: #Loop over each row in the column
            # according data description, the missing value is marked as -200
            if data == -200:
                nan_list[i] += 1
        print(f"{headers[i]:<13}    {nan_list[i]:>4}")
        
# Show missing value table
missing_search(headers)
```

```python
# From the missing table above, we see there are 366 rows with missing values in 9 of the columns
row_count = 0
for index, row in df.iterrows():
    inner_count = 0
    for header in headers:
        if row[header] == -200:
            inner_count += 1
    # There are more than 9 missing values out of 13 attribute
    if inner_count >= 9:
        row_count += 1
row_count
```

At this point, it is confirmed that there is 366 rows in the dataset having 9 missing values out of 13 attributes. We will delete these.

Also, there are no NaN and negative values.

On the next step, we plan to get rid of 'NMHC(GT)' column having too much missing data (8443/9357).

Finally, we will delete any remaining rows with missing values.

```python
# Drop column: NMHC(GT) due to its vast amount of missing data (8443/9357)
df.drop(['NMHC(GT)'], axis=1, inplace = True)
# Remove the column name from the list
headers.remove('NMHC(GT)')
# Then we delete the 366 rows discussed above. We will clear them based on the AH column.
df = df[df['AH'] != -200]
# Delete the rest of the rows with missing values, that are not part of the 366 rows discussed above.
df = df[df["CO(GT)"] != -200]
df = df[df["NOx(GT)"] != -200]
df = df[df["NO2(GT)"] != -200]
# Show missing value table again
missing_search(headers)
```

Next, we will confirm we still have a reasonable number of datapoints.

```python
df.shape
```

We have successfully cleaned our dataframe, and are able to keep 6941/9357 datapoints. 

We will now convert to a numpy array, but first the date and time attributes should be combined to the data type datetime.

```python
#First, combine the date and the time columns
df["Date_Time"] = df["Date"] +df["Time"]

#Convert
df['Date_Time'] =  pd.to_datetime(df['Date_Time'], format= '%d/%m/%Y%H.%M.%S')

#Remove old date and time columns
df.drop(labels=['Date','Time'], axis = 1, inplace = True)
headers.remove("Date")
headers.remove("Time")
headers[:0] = ["Date_Time"]
df.head()
```

Now, we have clean data and there is no missing values or invalid value. However, we are still need to get rid of PT08.S series data and just focus on one detector's data(GT).

So, we like to format the data into X vs y for high dimeneional regression analysis. NOx is one of the main gas for acid rain. And it's acidity is the strongest among other gas.

```python
interest_header = ["CO(GT)", "C6H6(GT)", "NO2(GT)", "T", "RH", "AH"]
X = df[interest_header]
X
```

```python
y = df["NOx(GT)"]
y
```

```python
X = np.array(X)
y = np.array(y)
y = y.reshape(-1, 1)
```

# Baseline Model# 

Let's plot the each X attribute to y to see their correlations

```python
fig, axes = plt.subplots(2, 3, figsize = (40,20))
axes = axes.ravel()
# plt.tick_params(axis='both', which='both', labelsize=22)
for i in range(len(interest_header)):
    axes[i].scatter(X[:, i], y)
    axes[i].tick_params(axis='both', labelsize=22)
    axes[i].set_xlabel(interest_header[i], fontsize=25)
    axes[i].set_ylabel('NOx(GT)', fontsize=25)
```
