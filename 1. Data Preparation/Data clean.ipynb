```python
%matplotlib inline
import numpy as np
import pandas as pd
import pylab as plt
```

# Importing data

```python
df = pd.read_csv('data/AirQualityUCI.csv', sep=';')
```

Print the labels to check if the data were loaded successfully. 

```python
for i in range(len(df.columns)):
    print(df.columns[i])
```

Some preliminary cleaning of the data set. 

```python
data = df.dropna(how='all')
data_clean = data.drop(columns=[df.columns[-1],df.columns[-2]])
print(data_clean)
```

Go through the data set to see if there are any NULL values left.

```python
data_bool = data_clean.isnull()
count = 0
num_row = len(data_bool[df.columns[0]])
num_col = len(data_bool.columns)
for i in range(num_col):
    for j in range(num_row):
        if data_bool[df.columns[i]][j]:
            print('column is: ', df.columns[i])
            print('row is: ', j)
            count = count + 1
print(count)
```

Print histogram of the each column to check the general distribution of the data set. 

```python
fig, axes = plt.subplots(5, 3, figsize = (20, 35), dpi = 150)

for i in range(2,num_col):
    i = i-2
    axes[int((i-i%3)/3),i%3].hist(data_clean[df.columns[i]],bins='auto')
    axes[int((i-i%3)/3),i%3].set_title('feature name: {}'.format(df.columns[i]))
```
