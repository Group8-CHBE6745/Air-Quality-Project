# Data Cleaning

```python
# Show the end of the df to see if there is any empty value
df.tail()
```

```python
# Drop end rows with NaN values
df.dropna(how = 'all', inplace = True)
# Show
df.tail()
```

```python
# Check if each attribute has the same count of data
print('Count of each values:\n', df.count())
```

From data decription, the missing value in the dataset is already replaced by -200. However, it is better to make sure there is no Nan and negative in the dataset.

```python
# Now, we are looking for the missing data in the dataframe
def missing_search(headers: list)->None:
    nan_list = [0] * 15
    print(f"Count of missing value out of {df.shape[0]} rows")
    for i in range(len(headers)):
        for data in df[headers[i]]:
            # according data description, the missing value is marked as -200
            if data == -200:
                nan_list[i] += 1
        print(f"{headers[i]:<13}    {nan_list[i]:>4}")
        
# Show missing value table
missing_search(headers)
```

```python
# Now, we are looking for the nagetive value and Nan
def invalid_check(headers: list)->None:
    nan_list = [0] * 15
    print(f"Count of Negative Value or NaN out of {df.shape[0]} rows")
    for i in range(len(headers)):
        for data in df[headers[i]]:
            # looking for the nagetive value and Nan except data and time
            if data == float and (data == None or data < 0):
                nan_list[i] += 1
        print(f"{headers[i]:<13}    {nan_list[i]:>4}")
invalid_check(headers)
```

```python
# From the missing table above, we see there are 366 rows with missing values
row_count = 0
for index, row in df.iterrows():
    inner_count = 0
    for header in headers:
        if row[header] == -200:
            inner_count += 1
    # There are more than 9 missing values out of 13 attribute
    if inner_count >= 9:
        row_count += 1
row_count
```

At this point, it is confirmed that there is 366 rows in the dataset having 9 missing values out of 13 attributes. Also, there is no NaN and negative values.
On the next step, we plan to get rid of 'NMHC(GT)' column having too much missing data (8443/9357) and then to delete the rows with missing values.

```python
# Drop column: NMHC(GT) due to its vast amount of missing data (8443/9357)
df.drop(['NMHC(GT)'], axis=1, inplace = True)
# Remove the column name from the list
headers.remove('NMHC(GT)')
# Then we delete 366 rows with a reference to AH (Could be T, RH, C6H6, etc.)
df = df[df['AH'] != -200]
# Delete the rest of the rows with missing values
df = df[df["CO(GT)"] != -200]
df = df[df["NOx(GT)"] != -200]
df = df[df["NO2(GT)"] != -200]
# Show missing value table again
missing_search(headers)
```
